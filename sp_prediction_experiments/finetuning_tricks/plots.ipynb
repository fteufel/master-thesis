```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Layer pruning

```python
df_sp = pd.read_csv('layer_pruning_xlnet/mcc_detection.csv')
df_cs = pd.read_csv('layer_pruning_xlnet/mcc_cs.csv')
df = df_sp.merge(df_cs)
```

```python
plt.figure(figsize=(9,4.5))
plt.subplot(1,2,1)
sns.scatterplot(x = 'remove_top_layers', y='Val Best mcc cs', data=df)
plt.ylabel('MCC Cleavage Site Detection', size=12)
plt.xlabel('Number of removed layers', size=12)
plt.ylim(0.5,1)

plt.subplot(1,2,2)
sns.scatterplot(x = 'remove_top_layers', y='Val Best mcc detection', data=df)
plt.ylabel('MCC Signal Peptide Detection', size=12)
plt.xlabel('Number of removed layers', size=12)
plt.ylim(0.8,1)
plt.tight_layout()

plt.savefig('xlnet_layer_pruning.png')
```

```python
sns.scatterplot(x='Val Best mcc detection', y='Val Best mcc cs', data=df, hue='remove_top_layers')
plt.xlim(0.9,1)
```
