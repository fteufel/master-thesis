```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('test_set_perplexities20-08-30-15_08.csv')
```

```python
df = df.rename(columns={'Unnamed: 0': 'Dataset'}).set_index('Dataset')
```

```python
df = df.rename(index={'custom_taxon_wasp': 'Parasitic Wasps'})
```

```python
sns.heatmap(df.T, cbar_kws={'label': 'Perplexity'}, annot=True, fmt='.3g')
plt.ylabel('Model')
plt.title('Model performances on taxon-specific datasets')
plt.show()
```
